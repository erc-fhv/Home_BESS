```python
from entsoe import EntsoePandasClient
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# Get the API key from environment variable
api_key = os.getenv("ENTSOE_API_KEY")
if api_key is None:
    raise ValueError("Please set the ENTSOE_API_KEY environment variable.")

# Readout day-ahead prices for Austria
client = EntsoePandasClient(api_key=api_key)
prices = client.query_day_ahead_prices(
    country_code="AT",
    start=pd.Timestamp("2025-01-01", tz="Europe/Berlin"),
    end=pd.Timestamp("2025-12-22", tz="Europe/Berlin"),
)

prices.to_csv("austria_day_ahead_prices.csv")
print(prices)
```

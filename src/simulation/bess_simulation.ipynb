```python
import pandas as pd
import bess
import pulp
import numpy as np
importlib.reload(bess)

my_bess = bess.Bess()
taeglicher_gewinn = []

# Jahressimulation
for act_date in pd.date_range("2025-01-01", "2025-12-23", tz="Europe/Vienna"):
    my_bess.run(act_date, use_dynamic_prices=True, verbose=False)
    gewinn = pulp.value(my_bess.pulp_model.objective)
    taeglicher_gewinn.append(gewinn)
    print(f"Gewinn für {act_date.date()}: {gewinn:.2f} EUR", flush=True)

print(f"Jahresgewinn: {np.sum(taeglicher_gewinn):.2f} EUR")
```

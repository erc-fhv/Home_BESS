```python
from pymodbus.client import ModbusTcpClient

IP = "10.0.0.9"
PORT = 1502
UNIT_ID = 1  # ggf. 1, 2 oder 247 testen

client = ModbusTcpClient(IP, port=PORT)

if not client.connect():
    print("❌ Verbindung fehlgeschlagen")
    exit()

print("✅ TCP Verbindung aufgebaut")

try:
    # SunSpec Common Model beginnt typischerweise bei 40000
    # pymodbus nutzt 0-basierte Adressen → 40001 = Adresse 0
    result = client.read_holding_registers(address=0, count=10, device_id=UNIT_ID)

    if result.isError():
        print("❌ Modbus-Fehler:", result)
    else:
        print("✅ Register gelesen:")
        print(result.registers)

finally:
    client.close()
```

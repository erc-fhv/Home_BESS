```python
import os
import requests
import datetime

API_KEY = os.environ["SOLAR_EDGE_API_KEY"]
SITE_ID = os.environ["SOLAR_EDGE_SITE_ID"]

# Zeitraum (max. 1 Monat)
end_time = datetime.datetime.now()
start_time = end_time - datetime.timedelta(days=30)

params = {
    "startTime": start_time.strftime("%Y-%m-%d %H:%M:%S"),
    "endTime": end_time.strftime("%Y-%m-%d %H:%M:%S"),
    "meters": "Production,Consumption,SelfConsumption,FeedIn,Purchased",
    "api_key": API_KEY
}

url = f"https://monitoringapi.solaredge.com/site/{SITE_ID}/powerDetails"
data = requests.get(url, params=params).json()["powerDetails"]

# =========================
# Get data into separate dictionaries
# =========================
production = {}
consumption = {}
self_consumption = {}
feed_in = {}
purchased = {}

for meter in data["meters"]:
    for v in meter["values"]:
        date = v["date"]
        value = v.get("value", 0)

        if meter["type"] == "Production":
            production[date] = value
        elif meter["type"] == "Consumption":
            consumption[date] = value
        elif meter["type"] == "SelfConsumption":
            self_consumption[date] = value
        elif meter["type"] == "FeedIn":
            feed_in[date] = value
        elif meter["type"] == "Purchased":
            purchased[date] = value

# =========================
# Ausgabe
# =========================
print(f"Zeitauflösung: {data['timeUnit']}, Einheit: {data['unit']}")
print("Datum                | PV | Verbrauch | Netz | Einspeisung | Eigenverbrauch")
print("-" * 80)

for date in sorted(production.keys()):
    print(
        f"{date} | "
        f"{production.get(date, 0):>5.0f} | "
        f"{consumption.get(date, 0):>9.0f} | "
        f"{purchased.get(date, 0):>4.0f} | "
        f"{feed_in.get(date, 0):>11.0f} | "
        f"{self_consumption.get(date, 0):>14.0f}"
    )
```
